**This notebook is written to collect required datasets and retain the neccessary information Data visualization
Datasets collected from the internet are :**

1.   India's vaccination details from COWIN's online API -   
https://api.cowin.gov.in/api/v1/reports/v2/getPublicReports?state_id=&district_id=
2.   State wise distribution of population -  
https://censusindia.gov.in/2011census/SC-ST/pca_state_distt_st.xls



Importing libraries to process the data

```python
import pandas as pd
import numpy as np
import urllib.request as urllib 
import json
```

Making API request and receiving the response .    
After receiving the response the json object     
from the response is stored in object named 'data'

```python
url='https://api.cowin.gov.in/api/v1/reports/v2/getPublicReports?state_id=&district_id='
response=urllib.urlopen(url,timeout=1)
data=json.loads(response.read())
data
```

Now for easier and faster  access to relevant data in the future visualization,we will store a part of the received data into a file name 'vaccinationDetails'. 

```python
vaccination_details=data['topBlock']['vaccination']
out=open('vaccinationDetails','w')
json.dump(vaccination_details,out,indent=6)
out.close()
```

```python
out=open('covid_vaccine_india','w')
json.dump(data,out,indent=6)
out.close()
```

The state wise vaccination details is extracted from the data object and converted to DataFrame .Finally its stored into csv file .

```python
stateWiseVaccination=pd.DataFrame(data['getBeneficiariesGroupBy'])
stateWiseVaccination.to_csv('stateWiseVaccination')
```

Now moving on to the second dataset ,i.e population dataset .
The dataset is loaded and stored in 'data2'

```python
data2=pd.read_csv('/content/stateWisePopulation.csv')
```

Using to masks to filter out only required data from the dataset and saving the dataset as csv file .

```python
statePopulation=data2[data2.Level==('State' or 'India')]
statePopulation=statePopulation[statePopulation.TRU=='Total']
statePopulation.to_csv('statePopulation')
```

Again using the Cowin data Api 

to collect past 3 months data regarding vaccination trend in India

Monthly report of April is appended to the date wise array

```python
date_wise=[]
for i in range(1,31):
  date='2021-04-'+str(i)
  url='https://api.cowin.gov.in/api/v1/reports/v2/getPublicReports?state_id=&district_id=&date='+date
  response=urllib.urlopen(url,timeout=1)
  daily=json.loads(response.read())
  date_wise.append(daily['topBlock']['vaccination']['total'])
```

Monthly report of May is appended to the date wise array

```python
for i in range(1,31):
  date='2021-05-'+str(i)
  url='https://api.cowin.gov.in/api/v1/reports/v2/getPublicReports?state_id=&district_id=&date='+date
  response=urllib.urlopen(url,timeout=1)
  daily=json.loads(response.read())
  date_wise.append(daily['topBlock']['vaccination']['total'])
```

Monthly report of June  is appended to the datewise array(upto 10th June)

```python
for i in range(1,11):
  date='2021-06-'+str(i)
  url='https://api.cowin.gov.in/api/v1/reports/v2/getPublicReports?state_id=&district_id=&date='+date
  response=urllib.urlopen(url,timeout=1)
  daily=json.loads(response.read())
  date_wise.append(daily['topBlock']['vaccination']['total'])
```

```python
dateWise=pd.DataFrame()
```

```python
dateWise['total']=date_wise
```

Saving the data as CSV for Visualizations 

```python
dateWise.to_csv('DateWiseTotal.csv')
```
